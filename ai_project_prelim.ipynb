```python
import pandas as pd
import numpy as np

# Load dataset
file_path = './annual-number-of-deaths-by-cause.csv'
data = pd.read_csv(file_path)

# Drop columns with more than 80% missing values
data = data.dropna(thresh=len(data) * 0.2, axis=1)

# Impute missing values for numerical columns
for col in data.columns:
    if data[col].dtype in ['float64', 'int64']:  # Only for numeric columns
        if data[col].isna().sum() / len(data) < 0.05:  # If <5% missing, fill with median
            data.loc[:, col] = data[col].fillna(data[col].median())  # Use .loc to avoid SettingWithCopyWarning
        elif data[col].isna().sum() / len(data) < 0.1:  # If <10% missing, fill with mean
            data.loc[:, col] = data[col].fillna(data[col].mean())  # Use .loc to avoid SettingWithCopyWarning

# Handle non-numeric columns (like 'Entity' and 'Year')
# If 'Entity' is relevant, encode it using One-Hot Encoding
data = pd.get_dummies(data, columns=['Entity'], drop_first=True)

# Verify cleaning process and types
print(data.info())
```

```python
import pandas as pd

# Load your dataset
file_path = './annual-number-of-deaths-by-cause.csv'
data = pd.read_csv(file_path)

# Define a function to determine focus diseases
def determine_focus_diseases(df):
    # Initialize a list to hold selected diseases
    selected_diseases = []

    # Define a threshold for minimum deaths to consider (you can adjust this as needed)
    min_deaths_threshold = 1000

    # Loop through columns (excluding non-numeric columns)
    for column in df.columns:
        if df[column].dtype in ['float64', 'int64']:
            # Calculate relevant statistics
            total_deaths = df[column].sum()
            max_deaths = df[column].max()
            mean_deaths = df[column].mean()
            count_non_zero = (df[column] > 0).sum()
            variance = df[column].var()

            # Apply selection criteria
            if total_deaths > min_deaths_threshold and variance > 0 and count_non_zero > 10:
                selected_diseases.append({
                    'Disease': column,
                    'Total Deaths': total_deaths,
                    'Max Deaths': max_deaths,
                    'Mean Deaths': mean_deaths,
                    'Count Non-Zero': count_non_zero,
                    'Variance': variance
                })

    # Convert selected diseases to a DataFrame for easier analysis
    selected_diseases_df = pd.DataFrame(selected_diseases)

    # Sort the DataFrame based on Total Deaths
    selected_diseases_df = selected_diseases_df.sort_values(by='Total Deaths', ascending=False)

    return selected_diseases_df

# Execute the function
focus_diseases_df = determine_focus_diseases(data)

# Print the resulting DataFrame
print("Diseases to focus on for modeling:")
print(focus_diseases_df)
```

```python
import pandas as pd

# Load the datasets
data = pd.read_csv('./annual-number-of-deaths-by-cause.csv')  # Adjust the sheet name if needed
population_data = pd.read_csv('world_population.csv')

# Check the columns of both DataFrames
print("Columns in main dataset:")
print(data.columns)

print("\nColumns in population dataset:")
print(population_data.columns)

# Handle potential missing columns
entity_column = 'Entity'
country_column = 'Country/Territory'

# Extract unique countries/entities from both datasets
try:
    unique_entities_data = set(data[entity_column].unique())
except KeyError:
    print(f"Column '{entity_column}' not found in the main dataset.")
    unique_entities_data = set()

try:
    unique_entities_population = set(population_data[country_column].unique())
except KeyError:
    print(f"Column '{country_column}' not found in the population dataset.")
    unique_entities_population = set()

# Identify mismatched entities
mismatched_entities = unique_entities_data - unique_entities_population
print("Mismatched Entities:", mismatched_entities)

# Remove mismatched entities from your main dataset
data_cleaned = data[~data[entity_column].isin(mismatched_entities)]

# Now you can merge the cleaned data with the population dataset
try:
    merged_data_cleaned = pd.merge(data_cleaned, population_data, left_on=[entity_column], right_on=[country_column], how='left')
except KeyError as e:
    print(f"Merge failed: {e}")

# Check for any remaining missing values in Population after cleaning
print("Remaining NaN values in Population after merge:", merged_data_cleaned['2022 Population'].isna().sum())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Load the cleaned dataset
data = pd.read_csv('./cleaned_annual-number-of-deaths-by-cause.csv')  # Adjust the path as needed

# Select features and target variable
features = data.drop(columns=['Entity', 'Year', 'Deaths - Neoplasms - Sex: Both - Age: All Ages (Number)'])
target = data['Deaths - Neoplasms - Sex: Both - Age: All Ages (Number)']

# Identify categorical and numerical columns
categorical_cols = features.select_dtypes(include=['object']).columns.tolist()
numerical_cols = features.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Create a preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numerical_cols),  # Impute numerical columns
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)  # One-Hot Encode categorical columns
    ])

# Create a pipeline with preprocessing and the model
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestRegressor(random_state=42))
])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Fit the model
pipeline.fit(X_train, y_train)

# Save the model for later use
import joblib
joblib.dump(pipeline, 'random_forest_model.pkl')
```

```python
from sklearn.model_selection import GridSearchCV

# Set up the parameter grid
param_grid = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [None, 10, 20],
    'model__min_samples_split': [2, 5, 10]
}

# Set up Grid Search with the pipeline
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, 
                           scoring='neg_mean_squared_error', cv=5, verbose=2)

# Fit Grid Search
grid_search.fit(X_train, y_train)

# Display the best parameters
print("Best parameters found: ", grid_search.best_params_)

# Save the best model
best_model = grid_search.best_estimator_
joblib.dump(best_model, 'best_random_forest_model.pkl')
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

# Load the best model
best_model = joblib.load('best_random_forest_model.pkl')

# Make predictions
y_pred = best_model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")

# Optional: Visualize predictions vs actual values
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual Deaths')
plt.ylabel('Predicted Deaths')
plt.title('Actual vs Predicted Deaths from Neoplasms')
plt.show()
```
